```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
import string
import math
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
```

```python
with open("doc_01.txt") as file:
  doc_1 = file.read()
```

```python
tokens_1 = nltk.word_tokenize(doc_1)
print(tokens_1)
```

```python
pos_tags = dict(zip(tokens_1, nltk.pos_tag(tokens_1)))
print(pos_tags)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()

tokens_1 = [lemmatizer.lemmatize(token) for token in tokens_1]

print(tokens_1)
```

```python
stemmer = nltk.stem.PorterStemmer()
tokens_1 = [stemmer.stem(token) for token in tokens_1]

tokens_1
```

```python
# stop word removal
def remove_stop_words( tokens ):
    return [ token for token in tokens if token not in nltk.corpus.stopwords.words('english') ]


tokens_1 = remove_stop_words(tokens_1)
print(tokens_1)
```

```python
def TF(tokens):
  leng = len(tokens)
  freq = dict((token, 0) for token in tokens)
  for token in tokens:
    freq[token]+= 1
  return dict([(token, count/leng) for token, count in freq.items()])


print(TF(tokens_1))

```

```python
all_tokens = nltk.word_tokenize(doc_1)

def calculate_idf():
    N = 1
    idf = dict()
    for token in all_tokens:
        f = 1
        if token in all_tokens:
            f+=1
        idf[token] = math.log(N/f)
    return idf


print(calculate_idf())
```
